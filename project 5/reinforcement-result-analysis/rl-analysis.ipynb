```python
import pandas as pd
import numpy as np
import ast
import seaborn as sns
#import renders as rs
from IPython.display import display # Allows the use of display() for DataFrames
import copy
import matplotlib.pyplot as plt

file_name= "simulation-statistics.csv"
df= pd.read_csv(file_name)


cgn_df= df[df["scenario"]=="cgn"]
lust_df= df[df["scenario"]=="lust"]

abs_cgn=cgn_df[cgn_df["clustering_algorithm"]=="clusters_maxabs_3dimensions"]

abs_cgn.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Emergency Stops"])
plt.show()
```

```python
rnd_cgn=cgn_df[cgn_df["clustering_algorithm"]=="random"]

rnd_cgn.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Emergency Stops","Teleports"])
plt.show()
```

```python
rnd_cgn=cgn_df[cgn_df["clustering_algorithm"]=="None"]

rnd_cgn.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Emergency Stops","Teleports"])
plt.show()
```

```python
abs_lust=lust_df[lust_df["clustering_algorithm"]=="clusters_maxabs_3dimensions"]

abs_lust.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Emergency Stops"])
plt.show()
```
