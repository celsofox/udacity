# CGN Result Metric Comparison

```python
import pandas as pd
import numpy as np
import ast
import seaborn as sns
#import renders as rs
from IPython.display import display # Allows the use of display() for DataFrames
import copy
import matplotlib.pyplot as plt

file_name= "simulation-statistics.csv"
df= pd.read_csv(file_name)


cgn_df= df[df["scenario"]=="cgn"]
lust_df= df[df["scenario"]=="lust"]




f, axes = plt.subplots(figsize=(25,16), nrows=3, ncols=3)
ax1,ax2, ax33=  axes[0]
ax3,ax4,ax55=  axes[1]
ax5,ax6, ax7=  axes[2]

ax1.set_title("cgn- reinforcement learner")
abs_cgn=cgn_df[cgn_df["clustering_algorithm"]=="clusters_maxabs_3dimensions"]
abs_cgn=abs_cgn[cgn_df["traffic_light_counts"]=="[1, 2, 3, 4]"]
abs_cgn.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Duration"], ax=ax1)

ax2.set_title("cgn- reinforcement learner")
ax2.set_ylim([0,1400])
abs_cgn.plot(x="simulation_steps", y=["Emergency Stops"], ax=ax2)

ax33.set_title("cgn- reinforcement learner")
abs_cgn.plot(x="simulation_steps", y=["DepartDelay"], ax=ax33)



ax3.set_title("cgn- random")
rnd_cgn=cgn_df[cgn_df["clustering_algorithm"]=="random"]
rnd_cgn.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Duration"], ax=ax3)

ax4.set_title("cgn- random")
rnd_cgn.plot(x="simulation_steps", y=["Emergency Stops"], ax=ax4)
ax55.set_title("cgn- random")
ax55.set_ylim([0,3])
rnd_cgn.plot(x="simulation_steps", y=["DepartDelay"], ax=ax55)


ax5.set_title("cgn- scenario controlled")
simu_cgn=cgn_df[cgn_df["clustering_algorithm"]=="None"]
simu_cgn=simu_cgn[simu_cgn["simulation_steps"]<4500]
simu_cgn.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Duration"], ax=ax5)

ax6.set_title("cgn- scenario controlled")
ax6.set_ylim([0,1400])
simu_cgn.plot(x="simulation_steps", y=["Emergency Stops"], ax=ax6)

ax7.set_title("cgn- scenario controlled")
ax7.set_ylim([0,3])
simu_cgn.plot(x="simulation_steps", y=["DepartDelay"], ax=ax7)


plt.savefig('cgn-simplereward.png', bbox_inches='tight')
plt.show()
```

# CGN Second Rewardfunction Metric- Comparison

```python
file_name= "simulation-statistics-2nd-reward.csv"
df= pd.read_csv(file_name)


cgn_df= df[df["scenario"]=="cgn"]




cgn_df=cgn_df.sort(['simulation_steps'])

f, axes = plt.subplots(figsize=(25,12), nrows=2, ncols=3)
ax1,ax2, ax33=  axes[0]
ax4,ax5, ax6=  axes[1]

ax1.set_title("cgn[1-4]- reinforcement learner")
abs_cgn=cgn_df[cgn_df["clustering_algorithm"]=="clusters_maxabs_3dimensions"]
abs_cgn=abs_cgn[cgn_df["traffic_light_counts"]=="[1, 2, 3, 4]"]
abs_cgn.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Duration"], ax=ax1)

ax2.set_title("cgn[1-4]- reinforcement learner")

abs_cgn.plot(x="simulation_steps", y=["Emergency Stops"], ax=ax2)

ax33.set_title("cgn[1-4]- reinforcement learner")
abs_cgn.plot(x="simulation_steps", y=["DepartDelay"], ax=ax33)


abs_cgn=cgn_df[cgn_df["traffic_light_counts"]=="[1, 2, 3, 4, 5, 6, 7]"]
ax4.set_title("cgn [1-7]- reinforcement learner")
abs_cgn.plot(x="simulation_steps", y=["WaitingTime","TimeLoss","Duration"], ax=ax4)


ax5.set_title("cgn [1-7]- reinforcement learner")

abs_cgn.plot(x="simulation_steps", y=["Emergency Stops"], ax=ax5)

ax6.set_title("cgn [1-7]- reinforcement learner")
abs_cgn.plot(x="simulation_steps", y=["DepartDelay"], ax=ax6)



plt.savefig('cgn-secondeward.png', bbox_inches='tight')
plt.show()
```
